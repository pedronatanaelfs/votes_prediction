```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_merged = pd.read_csv('../data/processed_Data/df_merged.csv')

df_merged
```

```python
# Create an empty DataFrame to store all filtered propositions
df_propositions = pd.DataFrame()

# Loop through the years 2000 to 2024
for year in range(2000, 2025):
    print(f"Reading file for year {year} ...")

    # Read the Excel file for the current year
    file_path = f"../data/propositions/proposicoes-{year}.xlsx"
    df_year = pd.read_excel(file_path)

    # Filter rows where 'id' in df_year matches 'proposicao_id' in df_merged
    df_year_filtered = df_year[df_year['id'].isin(df_merged['proposicao_id'])]
    
    print(f"Filtered {len(df_year_filtered)} rows from {len(df_year)} for year {year}.")

    # Concatenate the filtered rows to the master DataFrame
    df_propositions = pd.concat([df_propositions, df_year_filtered], ignore_index=True)

# Final status update
print("Finished reading and filtering all years.")
print(f"df_propositions shape: {df_propositions.shape}")
```

```python
df_propositions
```

```python
# 1. Group by 'proposicao_id' in df_merged to get the unique count of 'idvotacao'
grouped_votacoes = df_merged.groupby('proposicao_id')['idvotacao'].nunique()

# 2. Create a column in df_propositions for the unique voting session count
df_propositions['num_unique_voting_sessions'] = df_propositions['id'].map(grouped_votacoes)

# 3. (Optional) Fill any NaN values with 0 if some propositions do not appear in df_merged
df_propositions['num_unique_voting_sessions'] = df_propositions['num_unique_voting_sessions'].fillna(0)

# Print to check results
print(df_propositions[['id', 'num_unique_voting_sessions']].head())
```

```python
# 1. Get the counts of propositions for each number of voting sessions
counts = df_propositions['num_unique_voting_sessions'].value_counts().sort_index()

# 2. Create a bar chart
fig, ax = plt.subplots(figsize=(8, 4))

bars = ax.bar(
    counts.index.astype(str),  # Convert the index to string for labels on the x-axis
    counts.values,
    color='steelblue'
)

# 3. Remove grid lines
plt.grid(False)

# 4. Remove the y-axis
ax.yaxis.set_visible(False)

# 5. Annotate the bars with the counts above each bar
for bar in bars:
    height = bar.get_height()
    ax.annotate(
        f'{int(height)}',
        xy=(bar.get_x() + bar.get_width() / 2, height),
        xytext=(0, 3),  # Offset above the bar
        textcoords="offset points",
        ha='center', va='bottom'
    )

# Optional: remove the borders/spines for a cleaner look
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

# Add labels or a title if desired
ax.set_xlabel('Number of Voting Sessions')
ax.set_title('Distribution of Voting Sessions per Proposition')

plt.show()
```
